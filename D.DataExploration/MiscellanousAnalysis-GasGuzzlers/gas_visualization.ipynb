```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# Read the data from the file
with open('avg_gas_price_05-04-2023_16_10_39.txt', 'r') as file:
    data = eval(file.read())
```

```python
# Convert the data to a pandas dataframe
df = pd.DataFrame(data, columns=['Month-Year', 'Avg Gas Price'])

# Convert the 'Avg Gas Price' column to float
df['Avg Gas Price'] = df['Avg Gas Price'].astype(float)
```

```python
df
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(16, 8))

# Create a barplot with custom colors and style
barplot = sns.barplot(x='Month-Year', y='Avg Gas Price', data=df, palette='coolwarm')
barplot.set_title('Average Gas Price by Month-Year', fontsize=20, fontweight='bold')
barplot.set_xlabel('Month-Year', fontsize=16)
barplot.set_ylabel('Average Gas Price', fontsize=16)

# Add a line plot to the existing bar plot
lineplot = sns.lineplot(x='Month-Year', y='Avg Gas Price', data=df, color='red', marker='o', linewidth=2.5)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Customize the y-axis ticks format
barplot.yaxis.set_major_formatter(ticker.EngFormatter())

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

```python
# Convert Month-Year column to datetime
df['Month-Year'] = pd.to_datetime(df['Month-Year'], format='%m/%Y')

# Extract the year from the datetime column
df['Year'] = df['Month-Year'].dt.year

# Filter out data for year 2019
df = df[df['Year'] != 2019]

# Sort the data by year
df = df.sort_values(by=['Year'])

# Set the plot style and font scale
sns.set_style("darkgrid")
sns.set(font_scale=1.2)

# Create a 2x2 subplot grid for the remaining years (2015-2018)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

for i, year in enumerate(df['Year'].unique()):
    if year == 2019:
        continue
    row = i // 2
    col = i % 2
    subset = df[df['Year'] == year]
    ax = sns.lineplot(x='Month-Year', y='Avg Gas Price', color='red', data=subset, ax=axes[row, col])
    ax.set_title(f"Average Gas Prices for {year}", fontsize=16, fontweight='bold')
    ax.set_xlabel("Month-Year", fontsize=12, fontweight='bold')
    ax.set_ylabel("Average Gas Price", fontsize=12, fontweight='bold')
    ax.tick_params(axis='x', labelsize=10, rotation=45)
    ax.tick_params(axis='y', labelsize=10)

# Add space between the subplots
plt.subplots_adjust(hspace=0.5, wspace=0.3)   
plt.show()
```
